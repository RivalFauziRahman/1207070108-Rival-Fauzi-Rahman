```python
#import library
import numpy as np
import cv2
import matplotlib.pyplot as plt
import imageio
```

```python
img = imageio.imread("MU.jpg")
```

```python
#mendapatkan/define resolusi dan tipe gambar
img.height = img.shape[0]
img.width = img.shape[1]
img.channel = img.shape[2]
img.type= img.dtype
```

```python
#percobaan pertama kita buat brightness untuk gambar grayscale
#membuat variable img_brightness untuk menampung
img_brightness = np.zeros(img.shape, dtype = np.uint8)
```

```python
def brighter(nilai):
    for y in range(0,img.height):
        for x in range(0, img.width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red)+ int(green)+ int(blue))
            gray += nilai 
            if gray > 255:
                gray= 255
            if gray<0:
                gray = 0
            img_brightness[y][x] = (gray, gray, gray) 
```

```python
#menampilkan gambar dengan parameter -100
brighter (- 100)
plt.imshow(img_brightness)
plt.title('brighness -100')
plt.show()

brighter(25)
plt.imshow(img_brightness)
plt.title('brightness -100')
plt.show()
```

```python
#brightness RGB
img_rgbbrightness = np.zeros(img.shape, dtype=np.uint8)
```

```python
def rgbbrighter(nilai):
    for y in range(0, img.height):
        for x in range(0, img.width):
            red = img[y][x][0]
            red += nilai
            if red > 255:
                red = 255
            if red < 0:
                red = 0
            green = img[y][x][1]
            green += nilai
            if green > 255:
                green = 255
            if green < 0:
                green = 0
            blue = img[y][x][2]
            blue += nilai
            if blue > 255:
                blue = 255
            if blue < 0:
                blue = 0
            img_rgbbrightness[y][x] = (red, green, blue)
```

```python
rgbbrighter(-100)
plt.imshow(img_rgbbrightness)
plt.title("Brightness -100")
plt.show()

rgbbrighter(100)
plt.imshow(img_rgbbrightness)
plt.title("Brightness 100")
plt.show()
```

```python
#contrast
img_contrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
#fungsi untuk brightness RGB dengan nilai parameter
def contrass (nilai):
  for y in range(0, img.height):
    for x in range(0, img.width):
      red = img[y][x][0]
      green = img[y][x][1]
      blue = img[y][x][2]
      gray = (int(red)+int(green)+int(blue))/3
      gray += nilai
      if gray > 255 :
        gray = 255
      if gray < 0 :
        gray = 0
      img_contrass[y][x] = (gray, gray, gray)
```

```python
contrass(2)
plt.imshow(img_contrass)
plt.title("Contrass 2")
plt.show()

contrass(100)
plt.imshow(img_contrass)
plt.title("Contrass ")
plt.show()
```
